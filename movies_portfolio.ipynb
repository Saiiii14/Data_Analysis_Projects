```python
import pandas as pd 
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
import seaborn as sns 
from matplotlib.pyplot import figure 
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,8)
```

```python
df=pd.read_csv(r"C:\Users\Sai Sailesh\Downloads\movies_port.csv")
df
```

```python
df.dropna()
```

```python
df['yearcorrect'] = df['released'].astype(str).str.split().str[2]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['budget']=df['budget'].astype('int')
df['gross']=df['gross'].astype('int')
df['votes']=df['votes'].astype('int')
```

```python
df.head()
```

```python
df=df.sort_values(by='gross',inplace=False,ascending=False)
df
```

```python
df.drop_duplicates()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earinings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
plt.show()
```

```python
df1 = df.select_dtypes(include=[np.number])
df1
```

```python
df1.corr()
```

```python
df1.corr(method='pearson')
```

```python
df1.corr(method='spearman')
```

```python
correlation_matrix=df1.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True,annot_kws={"size":10})
plt.title("Correlation Matric for Numeric Features")
plt.ylabel("Budget for Film")
plt.xlabel("Movie Features")
plt.show()
```

```python
df.head()
```

```python
df_numerized=df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized
```

```python
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True,annot_kws={"size":8})
plt.title("Correlation Matric for Numeric Features")
plt.ylabel("Budget for Film")
plt.xlabel("Movie Features")
plt.show()
```

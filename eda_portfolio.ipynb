# what age groups are best in 50m race  ( 20 + racemin)

```python
# Correct usage of the query method
result = df3.query('race_length == "50mi"').groupby('athelete_age')['athelete_average_speed'].agg(['mean', 'count']).sort_values('mean',ascending=False).query('count>19')

# Display the result
print(result)
```

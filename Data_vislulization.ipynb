```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_stack=pd.read_csv("C:/Users/Sai Sailesh/Downloads/stacked_area_chart_data.csv")
df_stack
```

```python
colors=["#011638","#7e2987","#ef2026"]
labels=["Deisel","Petrol","Gas"]
sns.set_style("white")
plt.figure(figsize=(12,6))
plt.stackplot(df_stack["Year"],
              df_stack["Diesel"],
              df_stack["Petrol"],
              df_stack["Gas"],
              colors=colors,
              edgecolor='none')
plt.xticks(df_stack['Year'],rotation=45)
plt.legend(labels,loc="upper left")
plt.ylabel("Numbers of Cars",fontsize=13)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)",fontsize=13,fontweight='bold')
sns.despine()
plt.show()
```

```python
df_scatter=pd.read_csv("C:/Users/Sai Sailesh/Downloads/scatter_data.csv")
df_scatter
```

```python
plt.figure(figsize=(12,8))
scatter=plt.scatter(df_scatter['Area (ft.)'],
            df_scatter['Price'],alpha=0.6,
            c=df_scatter['Building Type'],
            cmap='viridis')
plt.legend(*scatter.legend_elements(),loc='upper left',title='Building Type')
plt.title("Relationship Between Arae and Price",fontsize=14,weight='bold')
plt.xlabel('Area in sq Feet', weight='bold')
plt.ylabel('Price in $',weight='bold')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=df_scatter['Price'],
                y=df_scatter['Area (ft.)'],
                hue=df_scatter['Building Type'],
                palette=['black','darkblue','pink','purple','red'],s=100)

plt.title("Relationship Between Arae and Price",fontsize=14,weight='bold')
plt.xlabel('Area in sq Feet', weight='bold')
plt.ylabel('Price in $',weight='bold')
plt.show()
```

```python
df_regression=pd.read_csv("C:/Users/Sai Sailesh/Downloads/scatter_plot_ii.csv")
df_regression
```

```python
sns.set(rc={'figure.figsize':(9,6)})
sns.regplot(x="Budget",
            y="Sales",
            data=df_regression,
            scatter_kws={'color':'k'},
            line_kws={'color':'red'})
plt.xlabel('Ad expenditure in dollars')
plt.ylabel('sales in dollars')
plt.title('Effect on Ad expenditure on sales',fontsize=14,weight='bold')

plt.show()
```

```python
sns.lmplot(x='Budget',
           y='Sales',
           data=df_regression,
           height=8,
           scatter_kws={'color':'k'},
           line_kws={'color':'red'})
plt.xlabel('Ad expenditure in dollars')
plt.ylabel('sales in dollars')
plt.title('Effect on Ad expenditure on sales',fontsize=14,weight='bold')
plt.show()
```

# Find the gender distributioin according to the target variable 

```python
data.head()
```

```python
data['sex'] = data['sex'].astype(str)
data['target'] = data['target'].astype(str)
sns.countplot(x='sex', hue='target', data=data)
plt.xticks([1,0],['Male','Female'])
plt.legend(loc='upper right',labels=['No disease','disease'])
plt.show()
```

# Check the age distribution in dataset 

```python
sns.histplot(x='age', bins=20, kde=True, color='blue', edgecolor='black', data=data)
plt.show()
```

# Check the test pain type 

```python
sns.countplot(x='cp',data=data)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal pain','asymptomatic'])
plt.show()
```

# Show the chest pain distribution according to the target variable 

```python
sns.countplot(x='cp',hue='target',data=data)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal pain','asymptomatic'])
plt.legend(loc='upper right',labels=['No disease','disease'])
plt.show()
```
